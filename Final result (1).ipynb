```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
```

```python
brands = []
Realprice = []
Discount = []
price = []
Load = []
Top = []
Color = []
RPM = []
Rating = []
for i in range(1,21):
    urls = """https://www.flipkart.com/search?q=washing+machines&sid=j9e%2Cabm%2C8qx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_sc_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_sc_na_na&as-pos=1&as-type=RECENT&suggestionId=washing+machines%7CWashing+Machines&requestId=fb2bc716-b407-444b-83f0-8a6fd41dba51&as-searchtext=washinf"""+str(i)
    print(urls)

    page = requests.get(urls)
    soup = BeautifulSoup(page.text)


    #Brand
    data=soup.find_all("div",class_="_4rR01T")
    for i in data:
        a = i.text
        brands.append(a.split()[0])


    #REAL_PRICE
    RP=soup.find_all("div",class_="_3I9_wc _27UcVY")

    for i in RP:
        a=i.text
        b=re.findall(r"â‚¹\d+,\d+",a)
        Realprice.append(b)



    #Discount
    discount=soup.find_all("div",class_="_3Ay6Sb")

    for i in discount[:24]:
        Discount.append(i.text)



    #PRICE
    p= soup.find_all("div",class_="_30jeq3 _1_WHN1")

    for i in p:
        price.append(i.text)



    #LOAD
    mass=soup.find_all("div",class_="_4rR01T")
    import numpy as np

    for i in mass:
        a=i.text
        b=re.findall(r"(?:6.5.kg|7.kg|7.5.kg|6.kg|8.kg|8.5.kg|9.kg)",a)
        if b:
            Load.append(b)
        else:
            Load.append(np.nan)



    #TOP_FRONT
    import numpy as np
    top=soup.find_all("div",class_="_4rR01T")

    for i in top:
        a = i.text
        b = re.findall(r"(?:Top|Front)",a)
        if b:
            Top.append(b)
        else:
            Top.append(np.nan)




    #COLOUR
    import numpy as np
    color=soup.find_all("div",class_="_4rR01T")

    for i in color:
        a = i.text
        b=re.findall(r"(?:Red|white|Grey|Silver|Maroon|Black|Purple)",a)
        if b:
            Color.append(b)
        else:
            Color.append(np.nan)




    #WHIRLING_SPEED
    speed= soup.find_all("li",class_="rgWa7D")

    for i in speed:
        a = i.text
        b = re.findall(r"(\d{3,}).(?:rpm|RPM)",a)
        if b:
            RPM.append(b)




    #RATING(ON-LINE)
    rating=soup.find_all("div",class_="_3LWZlK")

    for i in rating[:24]:
        Rating.append(i.text)
print(len(brands))
print(len(Realprice))
print(len(Discount))
print(len(price))
print(len(Load))
print(len(Top))
print(len(Color))
print(len(RPM))
print(len(Rating))
```

```python
d={"Brand":brands,
  "price":price,
  "Color":Color,
  "Load":Load,
  "Top_Front":Top,
   "RPM(maxspeed)":RPM,
  "discount":Discount,
  "Actualprice":Realprice,"Rating":Rating}
```
